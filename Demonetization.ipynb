```python
clean_data = data.drop(['messageID', 'Message Source', 'Profile Pic', 'Profile Link', 'From',
       'Bio', 'Title','Average Rating',
       'No. of Raters', 'Total views', 'Post Caption', 'Post Description',
       'Collected On','Timestamp',
       'Tags', 'Link', 'Source', 'From Name', 'From Email', 'From Team',
       'To Name', 'To Email', 'To Team', 'From Basket', 'To Basket',
       'Rule Name', 'Work Message', 'Action Start Date',
       'Action Date', 'Ticket ID', 'Notes', 'Date', 'Time', 'Served On',
       'Review'],axis=1)
```

```python
clean_data.info()
```

```python
clean_data.dropna(axis=0,inplace=True)
```

```python
clean_data.head()
```

```python
clean_data = clean_data.replace(' ',np.nan)
```

```python
round(clean_data.isnull().sum()/clean_data.shape[0]*100,2)
```
